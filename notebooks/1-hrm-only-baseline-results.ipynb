# MindTune Final Experimental Results

This notebook summarizes the final comparative evaluation of Random Forest and QSVM using custom HRM belt physiological data and selected feature sets.

```python
import pandas as pd

results = pd.DataFrame({
    "Experiment": [
        "A: hr_mean + hr_std + rmssd",
        "A: hr_mean + hr_std + rmssd",
        "B: hr_mean + hr_std + rmssd + hr_range",
        "B: hr_mean + hr_std + rmssd + hr_range"
    ],
    "Model": [
        "Random Forest",
        "QSVM",
        "Random Forest",
        "QSVM"
    ],
    "F1-score": [0.9906, 0.7876, 0.7901, 0.7263]
})

results
```

```python
import matplotlib.pyplot as plt

pivot = results.pivot(index="Experiment", columns="Model", values="F1-score")

pivot.plot(kind="bar", figsize=(10,6))
plt.title("Random Forest vs QSVM Across Feature Sets")
plt.ylabel("Average F1-score")
plt.ylim(0,1.05)
plt.xticks(rotation=15, ha="right")
plt.tight_layout()
plt.show()
```

# QUANTUM CIRCUIT

```python
%matplotlib inline

from qiskit.circuit.library import zz_feature_map
from qiskit.visualization import circuit_drawer
from IPython.display import display
import matplotlib.pyplot as plt
import os

os.makedirs("../report_figures", exist_ok=True)

qc = zz_feature_map(feature_dimension=3, reps=1)

fig = circuit_drawer(qc, output="mpl")
fig.savefig("../report_figures/figure_quantum_circuit.png", dpi=300, bbox_inches="tight")

display(fig)
plt.close(fig)

print("Saved to ../report_figures/figure_quantum_circuit.png")
```
